```python
import re

input_path = '/content/drive/My Drive/BOOK_DATA/books_batch_pages285_to_350.txt'
output_path = '/content/drive/My Drive/BOOK_DATA/cleaned_books_batch_pages285_to_350.txt'

def clean_line(line):
    line = line.lower()
    line = re.sub(r'[^a-z0-9\s.,;:!?\'"-]', '', line)
    line = re.sub(r'\s+', ' ', line).strip()
    return line

with open(input_path, 'r', encoding='utf-8', errors='ignore') as fin, \
     open(output_path, 'w', encoding='utf-8') as fout:
    for i, line in enumerate(fin):
        line = clean_line(line)
        if line:
            fout.write(line + '\n')
        if i % 50000 == 0:
            print(f"Processed {i} lines...")
```

```python
file_path = '/content/drive/My Drive/BOOK_DATA/books_batch_pages31_to_100.txt'

with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
    text = f.read()

print("Number of characters:", len(text))
print(text[:1000])  # show first 1000 characters
```

```python
import re
def clean_text(txt):
    # convert to lowercase
    txt = txt.lower()
    # remove special characters except letters, numbers, and punctuation
    txt = re.sub(r'[^a-z0-9\s.,;:!?\'"-]', '', txt)
    # remove extra spaces
    txt = re.sub(r'\s+', ' ', txt).strip()
    return txt

cleaned_text = clean_text(text)
print(cleaned_text[:1000])
```

```python
output_path = '/content/drive/My Drive/BOOK_DATA/cleaned_books_batch_pages31_to_100.txt'

with open(output_path, 'w', encoding='utf-8') as f:
    f.write(cleaned_text)

print("Cleaned text saved to:", output_path)
```
